## Customizing the Prompt and Model

Edit the two lines at the top of the script below — that's all you need to change:

| Variable | What to change |
|----------|----------------|
| `prompt` | Your question or instruction to the LLM |
| `model`  | Any model ID from [openrouter.ai/models](https://openrouter.ai/models) |

**Example prompts:**
```python
prompt = 'What is machine learning? Explain in one sentence.'
prompt = 'Translate "Hello, world!" into Japanese.'
prompt = 'Write a Python function that reverses a string.'
prompt = 'What are the differences between Dropout and Batch Normalization?'
```

**Example models:**
```python
model = 'google/gemini-2.0-flash-001'
model = 'anthropic/claude-3.5-haiku'
model = 'openai/gpt-4o-mini'
model = 'meta-llama/llama-3.3-70b-instruct'
```

> 📌 After editing, **run this cell first** (to save the script), then **run the next cell** (to execute it).

````python
# Save this cell as a Python file (Execute after editing)
%%writefile openrouter.py

import os
from openai import OpenAI

# ── Edit here ─────────────────────────────────────────────────────────────────
model  = 'google/gemini-2.0-flash-001'  # model ID from openrouter.ai/models
prompt = 'Explain what CIFAR-100 is in one sentence.'
# ──────────────────────────────────────────────────────────────────────────────

# Read API key from environment variable (set in the cell above)
API_KEY = os.environ.get('OPENROUTER_API_KEY')

# Create a client pointing to OpenRouter
client = OpenAI(
    base_url='https://openrouter.ai/api/v1',
    api_key=API_KEY,
)

# Send the prompt and print the response
response = client.chat.completions.create(
    model=model,
    messages=[{'role': 'user', 'content': prompt}],
)

print('Model :', model)
print('Prompt:', prompt)
print('Response:\n', response.choices[0].message.content)
````
