````python
# Step 2: Mount Google Drive and set up project folder
from google.colab import drive
drive.mount('/content/drive')

PROJECT_PATH="/content/drive/MyDrive/Practical-ML"
!mkdir -p "{PROJECT_PATH}"
%cd "{PROJECT_PATH}"
!ls *.py
````

````python
# Step 3: Edit your student ID, then run this cell
# %%writefile saves this code as studentid_hash.py into your Google Drive project folder
%%writefile studentid_hash.py

import hashlib

student_id = 'YOUR_STUDENT_ID'  # <- Change this to your student ID

digest = hashlib.sha256(student_id.encode()).digest()  # 32 bytes

# Fold all 32 bytes into 2 bytes by XOR (16 chunks of 2 bytes)
result = 0
for i in range(0, len(digest), 2):
    result ^= (digest[i] << 8) | digest[i+1]

print( f'Your student ID: {student_id}')
print( f'       You Hash: {result:04x}')
````
